```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

patient_data = pd.read_csv('healthcare-dataset-stroke-data.csv')

#let's have see how our dataset look like
patient_data 
```

```python
# To only execute the first few lines of dataset use this command
patient_data.head()
```

```python
# String will not be deal by our model therefore we eleminate the string type coulumns
edited_data_set = patient_data.drop(columns=['gender', 'ever_married', 'work_type', 'Residence_type','smoking_status','bmi'])

# let's see after deletion our dataset
edited_data_set
```

```python
X = edited_data_set.drop(columns=['stroke']) # this will create new dataset without having genre

y  = patient_data['stroke'] # In this we store value of our output data.
```

```python
# Data is stored in X without Stroke that is our input data
X.head()
```

```python
# we seperate the output data in y
y.head()
```

```python
y # whole dataset
```

```python
# let's devide the data into four different parts.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
percent = score
percent
```

```python
print("Accuracy is:", percent ,"%")# to read the output to see the in the percentage
```
